# Feature Engineering

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df['Day'],final_df['Month'],final_df['Year'] = (final_df['Date_of_Journey'].str.split('/').str)
# Alternate method
# final_df['Day'],final_df['Month'],final_df['Year'] = (final_df['Date_of_Journey'].apply(lambda x: x.split('/')))
```

```python
final_df.head()
```

```python
final_df['Day'] = final_df['Day'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis= 1, inplace= True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hour'],final_df['Arrival_min'] = (final_df['Arrival_Time'].apply(lambda x: x.split(':')).str)
```

```python
final_df['Arrival_hour'],final_df['Arrival_min'] = final_df['Arrival_hour'].astype(int),final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Dep_hour'],final_df['Dep_min'] = (final_df['Dep_Time'].apply(lambda x: x.split(':')).str)
final_df['Dep_hour'],final_df['Dep_min'] = final_df['Dep_hour'].astype(int),final_df['Dep_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Dep_Time','Arrival_Time'], axis= 1, inplace= True)
```

```python
final_df.head(3)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
```

```python
final_df['Total_Stops'].dropna(axis = 0)
final_df.isnull().sum()
```

```python
final_df.drop('Route', axis= 1, inplace= True)
```

```python
final_df['Additional_Info'].unique()
```

```python
# few flight have duration 5 min which is obviously not possible
final_df.drop(6474, inplace=True)
final_df.drop(2660, inplace=True)
```

```python
final_df[final_df['Duration'] == '5m']
```

```python
final_df['Duration'].isnull().sum()
```

```python
hr = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0].astype(int)
min = final_df['Duration'].str.split(' ').str[1]
# min = (final_df['|Duration'].str.split(' ')).str[1]
min.fillna('0', inplace= True)
```

```python
min = min.str.strip('m').astype(int)
```

```python
final_df['Duration_min'] = hr*60 + min
```

```python
final_df.drop('Duration',axis= 1, inplace= True)
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
